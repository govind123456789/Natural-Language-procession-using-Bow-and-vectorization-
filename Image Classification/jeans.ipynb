```python
from selenium import webdriver
import os
```

```python
driver = webdriver.Chrome(executable_path ='chromedriver.exe')
```

```python
def directory(dirname):
    current_path = os.getcwd()
    path = os.path.join(current_path, dirname)
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
directory("jeans")

```
